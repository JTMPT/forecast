מצב קיים

```python
matching_files=find_files_with_pattern(r'{}'.format(forecast_version_folder_location), '2020_jtmt_forcast_full')

# Get the newest file based on the timestamps
newest_file = max(matching_files, key=extract_timestamp)

file = pathlib.Path(newest_file)
file_name = file.name[:-5]

forecast_2020=up_load_df(forecast_version_folder_location, file_name)
```

מצב עתידי כל שנה

```python
file_date=pd.Timestamp.today().strftime('%y%m%d')

matching_files=find_files_with_pattern(r'{}\Intermediates'.format(cwd), '_pop_2040_jtmt')
file = pathlib.Path(matching_files[0])
v_date = file.name[:6]

year=['2025','2030','2035','2040','2045','2050']

for y in year: 
    folder_path=r'{}\Intermediates'.format(cwd)
    file_name='{}_forecast_{}_full'.format(v_date,y)

    locals()['forecast_{}'.format(y)]=up_load_df(folder_path,file_name)
```

```python
taz=up_load_shp(r'{}\background_files\TAZ_V4_{}_with_geo_info.shp'.format(create_forecast_basic_folder_path, TAZ_V4_date))

col_taz=['Taz_num',
 'Taz_name',
 'main_secto',
 'Muni_Heb',
 'jeru_metro',
 'zonetype',
 'in_jerusal',
 'SCHN_NAME',]
```

שלב ראשון קיצרתי את השמות שלהם

```python
for y in year: 
    locals()['forecast_{}'.format(y)]=locals()['forecast_{}'.format(y)][col_2050]
```

שלב שני לייצר לכל שנה רשימה חדשה של שמות השדות שלו עם שינוי

```python
for y in year: 
    col=['student_yeshiva',
         'student',
         'uni_students','aprt',
         'pop_without_dorms_yeshiva','emp_Education','emp_okev', 'emp_not_okev',
         'total_emp',
         'agri',
         'Indus',
         'Com_hotel',
         'Business',
         'Public']
    
    x=[]
    
    for i in col: x=x+[i+'_{}'.format(y)]
        
    x=['Taz_num']+x
    
    locals()['col_{}'.format(y)]=x
```

שלב שלישי לקחת את השמות החדשים ולשנות את העמודות של כל אחד

```python
for y in year: 
    locals()['forecast_{}'.format(y)].columns=locals()['col_{}'.format(y)]
```

שלב אחרון לחבר את כולם אל הקיים

```python
for y in year: 
    df=df.merge(locals()['forecast_{}'.format(y)],on='Taz_num',how='left')
```
